```python
import pandas as pd

df_ingredients = pd.read_csv("/content/_Ingredient__202410201632.csv")
df_recipe_ingredients = pd.read_csv("/content/_Recipe_ingredients__202410201631.csv")
df_recipe = pd.read_csv("/content/_Recipe__202410201554.csv")
```

```python
df_ingredients['ingredient_id'] = df_ingredients['id']
df_ingredients = df_ingredients.drop(["price","id_seq",'date_creation','date_update','type_ingredient_id','id'],axis = 1)

df_ingredients
```

```python
df_recipe_ingredients
```

```python
# Объединяем по 'id' ингредиента, чтобы получить названия ингредиентов
df_ri = df_recipe_ingredients.merge(df_ingredients, on='ingredient_id', how='left')
df_ri
```

```python
# Шаг 3: Создание индикаторной матрицы ингредиентов для каждого рецепта
ingredient_indicators = pd.crosstab(df_ri['recipe_id'], df_ri['name'])
# Шаг 4: Убедиться, что значения индикаторов являются 0 или 1
ingredient_indicators[ingredient_indicators > 0] = 1
ingredient_indicators
```

```python
df_recipe
```

```python
df_recipe['recipe_id'] = df_recipe['id']
df_recipe = df_recipe.drop(["id_seq",'url','calories','interactive_steps','common_steps','image','cook_time','author_id','date_creation','date_update','pfc','id','backend_ingredients'],axis = 1)

df_recipe
```

```python
df_recipe_with_ingredients = df_recipe.merge(ingredient_indicators, left_on='recipe_id', right_index=True, how='left')

df_recipe_with_ingredients
```

```python

# Преобразование 'is_meal' и 'is_drink' из строковых значений в числовые
df_recipe_with_ingredients['is_meal'] = df_recipe_with_ingredients['is_meal'].map({False: 0, True: 1})
df_recipe_with_ingredients['is_drink'] = df_recipe_with_ingredients['is_drink'].map({False: 0, True: 1})
df_recipe_with_ingredients
```

```python
# Список полей для one-hot encoding
cols_to_encode = ['spicyness', 'amount', 'temperature', 'alcoholicity', 'cuisine_id', 'drink_type_id', 'meal_type_id']

# Заполнение пропущенных значений (опционально, в зависимости от данных)
# Вы можете заменить пропущенные значения на специально обозначенное значение, например, 'unknown' или 'missing'
df_recipe_with_ingredients[cols_to_encode] = df_recipe_with_ingredients[cols_to_encode].fillna('unknown')

# Выполнение one-hot encoding
df_recipe_encoded = pd.get_dummies(df_recipe_with_ingredients, columns=cols_to_encode)
```

```python
bool_columns = [col for col in df_recipe_encoded.columns if df_recipe_encoded[col].dtype == (bool or float64)]
df_recipe_encoded[bool_columns] = df_recipe_encoded[bool_columns].astype(int)
```

```python
# Выбираем только числовые столбцы
df_recipes_numeric = df_recipe_encoded.select_dtypes(include=[np.number])

# Заполняем пропущенные значения нулями
df_recipes_numeric.fillna(0, inplace=True)

# Вычисляем матрицу косинусной близости
cosine_sim_matrix = pd.DataFrame(
    cosine_similarity(df_recipes_numeric),
    index=df_recipes_numeric.index,
    columns=df_recipes_numeric.index
)

def get_top_similar_recipes(recipe_id, cosine_sim_matrix, top_n=20):
    if recipe_id not in cosine_sim_matrix.index:
        return pd.DataFrame()
    similarity_series = cosine_sim_matrix.loc[recipe_id]
    sorted_similarities = similarity_series.sort_values(ascending=False)
    sorted_similarities = sorted_similarities.drop(recipe_id, errors='ignore')
    top_similarities = sorted_similarities.head(top_n)
    return top_similarities

# Пример использования функции
some_recipe_id = df_recipes_numeric.index[216]  # Замените на нужный ID рецепта
top_similar_recipes = get_top_similar_recipes(some_recipe_id, cosine_sim_matrix, top_n=20)

# Вывод списка ID похожих рецептов и значений косинусной близости
print(top_similar_recipes)
```

```python
class Get_Rec():
    def __init__(self, path_ingredients, path_recipe_ingredients, path_recipe):
        df_ingredients = pd.read_csv(path_ingredients)
        df_recipe_ingredients = pd.read_csv(path_recipe_ingredients)
        df_recipe = pd.read_csv(path_recipe)

        df_ingredients['ingredient_id'] = df_ingredients['id']
        df_ingredients = df_ingredients.drop(["price","id_seq",'date_creation','date_update','type_ingredient_id','id'],axis = 1)
        df_ri = df_recipe_ingredients.merge(df_ingredients, on='ingredient_id', how='left')

        ingredient_indicators = pd.crosstab(df_ri['recipe_id'], df_ri['name'])
        ingredient_indicators[ingredient_indicators > 0] = 1

        df_recipe['recipe_id'] = df_recipe['id']
        df_recipe = df_recipe.drop(["id_seq",'url','calories','interactive_steps','common_steps','image','cook_time','author_id','date_creation','date_update','pfc','id','backend_ingredients'],axis = 1)
        df_recipe_with_ingredients = df_recipe.merge(ingredient_indicators, left_on='recipe_id', right_index=True, how='left')

        df_recipe_with_ingredients['is_meal'] = df_recipe_with_ingredients['is_meal'].map({False: 0, True: 1})
        df_recipe_with_ingredients['is_drink'] = df_recipe_with_ingredients['is_drink'].map({False: 0, True: 1})

        cols_to_encode = ['spicyness', 'amount', 'temperature', 'alcoholicity', 'cuisine_id', 'drink_type_id', 'meal_type_id']
        df_recipe_with_ingredients[cols_to_encode] = df_recipe_with_ingredients[cols_to_encode].fillna('unknown')
        df_recipe_encoded = pd.get_dummies(df_recipe_with_ingredients, columns=cols_to_encode)

        bool_columns = [col for col in df_recipe_encoded.columns if df_recipe_encoded[col].dtype == (bool or float64)]
        df_recipe_encoded[bool_columns] = df_recipe_encoded[bool_columns].astype(int)

        df_recipe_encoded = df_recipe_encoded.fillna(0)

        float_cols = df_recipe_encoded.select_dtypes(include='float64').columns
        df_recipe_encoded[float_cols] = df_recipe_encoded[float_cols].astype(int)

        if 'recipe_id' in df_recipe_encoded.columns:
            df_recipe_encoded.set_index('recipe_id', inplace=True)

        self.df_recipes_numeric = df_recipe_encoded.select_dtypes(include=[np.number])
        self.df_recipes_numeric.fillna(0, inplace=True)

        # Вычисляем матрицу косинусной близости
        self.cosine_sim_matrix = pd.DataFrame(
            cosine_similarity(self.df_recipes_numeric),
            index=self.df_recipes_numeric.index,
            columns=self.df_recipes_numeric.index
        )

    def get_top_similar_recipes(self, recipe_id, top_n=20):
        # recipe_id = self.df_recipes_numeric.index[recipe_id]
        if recipe_id not in self.cosine_sim_matrix.index:
            return pd.DataFrame()

        similarity_series = self.cosine_sim_matrix.loc[recipe_id]
        sorted_similarities = similarity_series.sort_values(ascending=False)
        sorted_similarities = sorted_similarities.drop(recipe_id, errors='ignore')
        top_similarities = sorted_similarities.head(top_n)

        return top_similarities
```
